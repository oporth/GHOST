```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = '../../2D/bin/run/'
!ls '../../2D/bin/run/'
shape = (1024,1024)
```

```python
offset = 10
ps = np.fromfile(path+'ps.'+str(offset).zfill(3)+'.out',dtype=np.float32).reshape(shape,order='F')
az = np.fromfile(path+'az.'+str(offset).zfill(3)+'.out',dtype=np.float32).reshape(shape,order='F')
```

```python
fig=plt.figure()
im=plt.imshow(az,origin='lower')
ax=plt.gca()
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Az, it={}'.format(offset))

fig=plt.figure()
im=plt.imshow(ps,origin='lower')
ax=plt.gca()
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Ps, it={}'.format(offset))
```
